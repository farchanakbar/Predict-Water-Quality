## Splitting Data

```python
X = data.drop('is_safe', axis=1)  # for feature
y = data['is_safe'] # for target
```

```python
# 5 first rows of the feature
X.head()
```

```python
# 5 first rows of target
y.head().to_frame()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# use 20% for data test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# check shape data train and test
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# fit and transform data train and test with standard scaler
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Create Model

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
# name variable of the model
svc = SVC()
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier(n_estimators=200)
gbc = GradientBoostingClassifier()
```

```python
# create a function for fit and predict model and report predict
def model(model):
    model.fit(X_train_scaled,y_train)
    y_pred = model.predict(X_test_scaled)
    print(classification_report(y_test,y_pred))
    plot_confusion_matrix(model, X_test_scaled, y_test);
```

```python
# prediction report from the model support vector machine
print('Support Vectore Machine')
print('\n')
model(svc)
```

```python
# prediction report from the model DecisionTreeClassifier
print('DecisionTreeClassifier')
print('\n')
model(dtc)
```

```python
# prediction report from the model RandomForestClassifier
print('RandomForestClassifier')
print('\n')
model(rfc)
```

```python
# prediction report from the model GradientBoostingClassifier
print('GradientBoostingClassifier')
model(gbc)
```

## Using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# parameter of GradientBoostingClassifier
param_grid = {"n_estimators":[1,5,10,20,40,100],'max_depth':[3,4,5,6]}
```

```python
grid = GridSearchCV(gbc,param_grid, verbose=2)
```

```python
model(grid)
```
